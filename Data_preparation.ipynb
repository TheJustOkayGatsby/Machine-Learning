```python
####### Data Preparation #######################
# reading the dataset
from pandas import read_csv, get_dummies,DataFrame,Series
dataset=read_csv('/content/drive/MyDrive/Colab Notebooks/Emloyees.csv')
# dataset
# dataset.info()
##################################################################################
# Data encoding
dataset['PastEmployee'] = dataset['PastEmployee'].map({'Yes':1, 'No':0})
dataset['OverTime'] = dataset['OverTime'].map({'Yes':1, 'No':0})

dataset['Gender'] = dataset['Gender'].map({'Female':1, 'Male':0})
# print(dataset.info())
# dataset.head(2)
#####
categorical_features = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
final_data = get_dummies(dataset, columns = categorical_features, drop_first=True)
# print(final_data.info())
# print(final_data.head(2))
##################################################################################
#Determine the input feature and output label
X = final_data.drop('PastEmployee', axis = 1) # Features
Y = final_data['PastEmployee'] # Labels
# # print(type(X))
# # print(type(Y))
# print(X.shape)
# print(Y.shape)
################################################################################### Data Scaling (data normalization)
from sklearn.preprocessing import StandardScaler
feature_scaler = StandardScaler()
X_scaled = feature_scaler.fit_transform(X)
#print (X_scaled)
# print (DataFrame(X_scaled).head())
################################################################################### Data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X_scaled, Y, test_size = 0.3, random_state = 100)
# print(X_train.shape)
# print(X_test.shape)
################################################################################### Data balancing
from imblearn.over_sampling import SMOTE
print("Number of observations in each class before oversampling (training data): \n", Series(Y_train).value_counts())
smote = SMOTE(random_state = 101)
X_train,Y_train = smote.fit_resample(X_train,Y_train)
print("Number of observations in each class after oversampling (training data): \n", Series(Y_train).value_counts())

DataFrame(X_train)
```
